# 8 пункт

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.scatter(data["Age"], data["y"], s=50, facecolor='C0', edgecolor='k')
ax.set_xlabel('Возраст')
ax.set_ylabel('целевое значение')
```

```python
df_exploded = data.assign(ReasonsForLack=data['ReasonsForLack'].str.split(', ')).explode('ReasonsForLack')
df_exploded = df_exploded.fillna(value="NoReason")
grouped_data = df_exploded.groupby('ReasonsForLack')['y'].mean().reset_index()
grouped_data = grouped_data.sort_values(by='y', ascending=False)

plt.figure(figsize=(12, 8))
plt.bar(grouped_data['ReasonsForLack'], grouped_data['y'], color='skyblue')
plt.xlabel('Reasons for Lack')
plt.ylabel('Average y')
plt.title('Average y by Reasons for Lack')
plt.ylim(0, 1)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
