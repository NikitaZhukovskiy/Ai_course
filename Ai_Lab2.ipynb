# 8 пункт

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(hue='y', y='Age', data=data, palette='Set2')

plt.xlabel('y')
plt.ylabel('Age')
```

```python
df = pd.DataFrame(data)

heatmap_data = df.groupby(['SelfConfidence', 'y']).size().unstack(fill_value=0)

plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap="crest")

plt.title('Тепловая карта зависимости y от SelfConfidence')

plt.show()
```

```python
df = pd.DataFrame(data)

heatmap_data = df.groupby(['Competitiveness', 'y']).size().unstack(fill_value=0)

plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap="crest")

plt.title('Тепловая карта зависимости y от Competitiveness')
```

```python
df = pd.DataFrame(data)

heatmap_data = df.groupby(['StrongNeedToAchieve', 'y']).size().unstack(fill_value=0)

plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap="crest")

plt.title('Тепловая карта зависимости y от StrongNeedToAchieve')
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='GoodPhysicalHealth', hue='y', fill=True, common_norm=False)

plt.xlabel('GoodPhysicalHealth')
plt.ylabel('Density')
```
